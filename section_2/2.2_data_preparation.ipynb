# 2.2 Data preparation

```python
data = 'https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-02-car-price/data.csv'
```

```python
df = pd.read_csv(data)
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_') # makes everything lowercase and replaces spaces with underscores
```

```python
df.head()
```

```python
df.dtypes
```

```python
strings = list(df.dtypes[df.dtypes == 'object'].index)
strings
```

```python
for col in strings: # col = column
    df[col] = df[col].str.lower().str.replace(' ', '_')
```
