# 3.6 - Feature importance: mutual information

Mutual information: concept from information theory, it tells us how much we can learn aboute one variable if we know the value of another

```python
from sklearn.metrics import mutual_info_score
```

```python
mutual_info_score(df_full_train.churn, df_full_train.contract)

# this tells how much we learn about churn, by absorbing the contract variable and vice-versa
```

```python
mutual_info_score(df_full_train.contract, df_full_train.churn) # if we invert the inputs, the outcome is the same
```

```python
mutual_info_score(df_full_train.churn, df_full_train.gender)

# gender doesn't tell us much about the churn
```

```python
mutual_info_score(df_full_train.churn, df_full_train.partner)

# partner is more important than gender, but far less than contract
```

```python
# Doing this for all the categorical variables:

def mutual_info_churn_score(series):
    return mutual_info_score(series, df_full_train.churn)
```

```python
# Mutual info scores for all categorical vars:

mi = df_full_train[categorical].apply(mutual_info_churn_score)

mi.sort_values(ascending=False)
```
